```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_original = cv2.imread("./peppers.png",1)
```

```python
plt.imshow(img_original)
```

```python
img_rgb = cv2.cvtColor(img_original, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img_original, cv2.COLOR_BGR2GRAY)
```

```python
fig = plt.figure(figsize=(10, 7)) 
  
# setting values to rows and column variables 
rows = 2
columns = 2


# Adds a subplot at the 1st position 
fig.add_subplot(rows, columns, 1) 
plt.imshow(img_original) 
plt.axis('off') 
plt.title("BGR Image") 
  
# Adds a subplot at the 2nd position 
fig.add_subplot(rows, columns, 2) 
plt.imshow(img_rgb) 
plt.axis('off') 
plt.title("RGB Image") 
  
# Adds a subplot at the 3rd position 
fig.add_subplot(rows, columns, 3) 
plt.imshow(img_gray, cmap="gray") 
plt.axis('off') 
plt.title("Gray") 
```

```python
#RGB'yi HSV'ye dntr
img_hsv_green = cv2.cvtColor(img_original, cv2.COLOR_BGR2HSV)

#Yeil iin HSV araln belirleyin
lower_green = np.array([20, 100, 90]) #yeil iin alt HSV snr
upper_green = np.array([130, 210,130]) #yeil iin st HSV snr

#yeil renkleri maskelemek iin aralklar uygulayn
mask_green = cv2.inRange(img_hsv_green, lower_green, upper_green)
```

```python
fig = plt.figure(figsize=(10, 7)) 
  
# setting values to rows and column variables 
rows = 1
columns = 2

# Adds a subplot at the 1st position 
fig.add_subplot(rows, columns, 1) 
plt.imshow(img_rgb_new_1) 
plt.axis('off') 
plt.title("RGB Image") 
  
# Adds a subplot at the 3rd position 
fig.add_subplot(rows, columns, 2) 
plt.imshow(mask_green, cmap="gray") 
plt.axis('off') 
plt.title("indexes image") 
```

```python
#kodlarla renk oluturma

height, width = 100, 100
#Krmz renk oluturma
img_red = np.zeros((height, width,3),dtype=np.uint8)
img_red[:,:,0] = 255
#Yeil renk oluturma
img_green = np.zeros((height, width,3),dtype=np.uint8)
img_green[:,:,1] = 255
#Mavi renk oluturma
img_blue = np.zeros((height, width,3),dtype=np.uint8)
img_blue[:,:,2] = 255
#Mavi renk oluturma
img_yellow = np.zeros((height, width,3),dtype=np.uint8)
img_yellow[:,:,:2] = 255
```

```python
fig = plt.figure(figsize=(10, 7)) 
  
# setting values to rows and column variables 
rows = 2
columns = 4

# Adds a subplot at the 1st position 
fig.add_subplot(rows, columns, 1) 
plt.imshow(img_red) 
plt.axis('off') 
plt.title("Red Image") 
  
# Adds a subplot at the 3rd position 
fig.add_subplot(rows, columns, 2) 
plt.imshow(img_green, cmap="gray") 
plt.axis('off') 
plt.title("green image") 
  
# Adds a subplot at the 3rd position 
fig.add_subplot(rows, columns, 3) 
plt.imshow(img_blue, cmap="gray") 
plt.axis('off') 
plt.title("blue image") 

# Adds a subplot at the 3rd position 
fig.add_subplot(rows, columns, 4) 
plt.imshow(img_yellow, cmap="gray") 
plt.axis('off') 
plt.title("yellow image") 
```

```python
img_black = np.zeros((height, width,3),dtype=np.uint8)

img_white = np.ones((height, width,3),dtype=np.uint8)*255
```

```python
fig = plt.figure(figsize=(10, 7), facecolor='purple') 
  
# setting values to rows and column variables 
rows = 2
columns = 4

# Adds a subplot at the 1st position 
fig.add_subplot(rows, columns, 1) 
plt.imshow(img_black) 
plt.axis('off') 
plt.title("Black Image") 

# Adds a subplot at the 1st position 
fig.add_subplot(rows, columns, 2) 
plt.imshow(img_white) 
plt.axis('off') 
plt.title("White Image") 
  
```
