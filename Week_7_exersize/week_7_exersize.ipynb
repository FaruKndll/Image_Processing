```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# soru 2 görüntünüzü gri seviyede tek kanallı olarak okuyunuz ve bunları bir değişkende saklayınız.
# saru 3 Bu görüntüyü tek bir plot içereisinde yan yana gösteriniz. Başlık bilgilerinde renk formatlarını yazınız.
```

```python
img_original = cv2.imread("./cat.jpg",1)
```

```python
img_rgb = cv2.cvtColor(img_original, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img_original, cv2.COLOR_BGR2GRAY)
```

```python
fig = plt.figure(figsize=(10, 7)) 
  
# setting values to rows and column variables 
rows = 1
columns = 3


# Adds a subplot at the 1st position 
fig.add_subplot(rows, columns, 1) 
plt.imshow(img_original) 
plt.axis('off') 
plt.title("BGR Image") 
  
# Adds a subplot at the 2nd position 
fig.add_subplot(rows, columns, 2) 
plt.imshow(img_rgb) 
plt.axis('off') 
plt.title("RGB Image") 
  
# Adds a subplot at the 3rd position 
fig.add_subplot(rows, columns, 3) 
plt.imshow(img_gray, cmap="gray") 
plt.axis('off') 
plt.title("Gray")
```
