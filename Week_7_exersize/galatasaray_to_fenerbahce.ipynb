```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Grnty ykleyelim (BGR formatnda gelir)
img_original = cv2.imread("./salatasaray.jpg", 1)  # Grnt yolu ile deitirin

# BGR'yi RGB'ye dntrelim (OpenCV BGR okur, matplotlib RGB bekler)
img_rgb = cv2.cvtColor(img_original, cv2.COLOR_BGR2RGB)

# Krmz renkli piksellerin maskesini oluturuyoruz (Krmz genellikle [255, 0, 0] olur)
# Ancak biraz daha esnek olabilmesi iin koulda tolerans kullanyoruz
mask_red = (img_rgb[:,:,0] > 150) & (img_rgb[:,:,1] < 100) & (img_rgb[:,:,2] < 100)

# Krmz piksellere mavi renk (RGB = [0, 0, 255]) atyoruz
img_rgb[mask_red] = [0, 0, 255]

# Sonular grselletirelim
fig = plt.figure(figsize=(10, 7)) 

# 1. Subplot: Deitirilen grnt
fig.add_subplot(1, 1, 1) 
plt.imshow(img_rgb) 
plt.axis('off') 
plt.title("Red to Blue Transformation") 
plt.show()

# Deitirilmi resmi kaydedelim
cv2.imwrite("modified_image.jpg", cv2.cvtColor(img_rgb, cv2.COLOR_RGB2BGR))  # BGR formatna geri evirip kaydediyoruz
```
