```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Görüntüyü yükleyelim (BGR formatında gelir)
img_original = cv2.imread("./salatasaray.jpg", 1)  # Görüntü yolu ile değiştirin

# BGR'yi RGB'ye dönüştürelim (OpenCV BGR okur, matplotlib RGB bekler)
img_rgb = cv2.cvtColor(img_original, cv2.COLOR_BGR2RGB)

# Kırmızı renkli piksellerin maskesini oluşturuyoruz (Kırmızı genellikle [255, 0, 0] olur)
# Ancak biraz daha esnek olabilmesi için koşulda tolerans kullanıyoruz
mask_red = (img_rgb[:,:,0] > 150) & (img_rgb[:,:,1] < 100) & (img_rgb[:,:,2] < 100)

# Kırmızı piksellere mavi renk (RGB = [0, 0, 255]) atıyoruz
img_rgb[mask_red] = [0, 0, 255]

# Sonuçları görselleştirelim
fig = plt.figure(figsize=(10, 7)) 

# 1. Subplot: Değiştirilen görüntü
fig.add_subplot(1, 1, 1) 
plt.imshow(img_rgb) 
plt.axis('off') 
plt.title("Red to Blue Transformation") 
plt.show()

# Değiştirilmiş resmi kaydedelim
cv2.imwrite("modified_image.jpg", cv2.cvtColor(img_rgb, cv2.COLOR_RGB2BGR))  # BGR formatına geri çevirip kaydediyoruz
```
